```python
# Convert data to the list of lists
transactions = []
for i in range(7501):
    transactions.append([str(data.values[i, j]) for j in range(20)])
```

```python
# Training the data
# First find the products that was at least purchased 3 times a day: support = 3*7/7500 = 0.0028
# (transactions was registered a week)
# confidence 
from apyori import apriori
rules = apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, max_length=2)
```

```python
results = list(rules)
```
